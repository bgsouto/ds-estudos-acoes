```python
#instalação dos modulos utilizados

!pip install yfinance --upgrade --no-cache-dir
!pip install -U plotly

#Importação das bibliotecas necessárias

import yfinance as yf
import pandas as pd
import plotly.express as px
from google.colab.data_table import DataTable
```

```python
#Importando lista do S&P 500 da wikipedia (ja contendo subsetores e setores de acordo com o Global Industry Classification Standard GICS)

url ='https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp500 = pd.read_html(url)[0]
sp500.sort_values('Symbol', ascending=True)

```

```python
#Importando a lista de pesos do S&P 500 do site slickcharts, é necessário usar o método requests para simular a requisição de um navegador, senão o site barra a requisição por considerar que é um robô scrapper.

import requests
url2 ='https://www.slickcharts.com/sp500'
sp500peso = pd.read_html(requests.get('https://www.slickcharts.com/sp500',headers={'User-agent': 'Mozilla/5.0'}).text)[0]
sp500peso.sort_values('Symbol', ascending=True)
```

```python
#juntando os 2 data frames para termos a informação completa do peso de cada ação em apenas um dataframe
sp500["Weight"] = sp500.merge(sp500peso, on="Symbol")["Weight"]
```

```python
sp500.sort_values('Symbol', ascending=True)
```

```python
#configurando a figura de um grafico sunburst utilizando o plotly e ajustando para mostrar de acordo com os pesos.

figura = px.sunburst(data_frame=sp500, path=['GICS Sector', 'GICS Sub-Industry', 'Symbol'], values='Weight', height=1200)

figura.update_layout(
    title_text='S&P 500 Peso por Setor, Sub-Setor e Stock',
    title_font_family="Arial",
    title_font_color="Black",
    title_x=0.5)

figura.write_html("SeP500_Sunburst.html")

figura.update_traces(textfont_color='white',
                  textfont_size=14,
                  hovertemplate='<b>%{label}:</b> %{value:.2f}%')


figura.show()
```

```python
#configurando a figura de um grafico treemap utilizando o plotly e ajustando para mostrar de acordo com os pesos.

figura = px.treemap(data_frame=sp500, path=['GICS Sector', 'GICS Sub-Industry', 'Symbol'], values='Weight', height=1200)

figura.update_traces(textfont_color='white',
                  textfont_size=14,
                  hovertemplate='<b>%{label}:</b> %{value:.2f}%')

figura.update_layout(
    title_text='S&P 500 Peso por Setor, Sub-Setor e Stock',
    title_font_family="Arial",
    title_font_color="Black",
    title_x=0.5)

figura.write_html("SeP500_Treemap.html")

figura.show()
```
