```python
#instalação dos modulos utilizados

!pip install yfinance --upgrade --no-cache-dir
!pip install -U plotly

#Importação das bibliotecas necessárias

import yfinance as yf
import pandas as pd
import plotly.express as px
from google.colab.data_table import DataTable
```

```python
#Importando lista de todos US REITs


import requests
url ='https://stockmarketmba.com/listofreits.php'
#usreits = pd.read_html(requests.get(url,headers={'User-agent': 'Mozilla/5.0'}).text)[0][:-1]
usreits = pd.read_html(url, thousands=',')[0][:-1].copy()
usreits
```

```python
#usreits['Industry'].apply(lambda s: s[s.rfind("�")+1:])
usreits.loc[:,'Industry'].apply(lambda s: s[s.rfind("�")+1:])
```

```python
#tirando o "REIT" da categoria industrial e limpando o campo
#usreits['Industry'] = usreits['Industry'].apply(lambda s: s[s.rfind("")+1:])
usreits.loc[:,'Industry'] = usreits['Industry'].apply(lambda s: s[s.rfind("�")+1:])
usreits.loc[:,'Market cap'] = usreits.loc[:,'Market cap'].apply(lambda s: s[s.rfind("$")+1:])


#remove as "," do campo do market e cap e depois transforma o campo em um valor numerico float
usreits.loc[:,'Market cap'] = usreits.loc[:,'Market cap'].str.replace(',','')
usreits.loc[:,'Market cap'] = pd.to_numeric(usreits.loc[:,'Market cap'], downcast="float")

usreits
```

```python
usreits
```

```python
#configurando a figura de um grafico sunburst utilizando o plotly e ajustando para mostrar de acordo com os pesos.

figura = px.sunburst(data_frame=usreits, path=['Industry', 'Symbol'], values='Market cap', height=1200)

figura = figura.update_traces(textfont_color='white',
                  textfont_size=14,
                  hovertemplate='<b>%{label}</b> - MktCap: $%{value:,.0f}')

figura.update_layout(
    title_text='US REITs - Grafico de Market Cap por Setor',
    title_font_family="Arial",
    title_font_color="Black",
    title_x=0.5)

figura.write_html("REITs_Sunburst.html")

figura.show()
```
